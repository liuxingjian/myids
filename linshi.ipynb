```python
import pandas as pd

df = pd.read_csv('flows.0.csv')
df = df[~df.apply(lambda row: row.astype(str).str.contains('SSH', na=False).any(), axis=1)]
df = df[~df.apply(lambda row: row.astype(str).str.contains('DNS', na=False).any(), axis=1)]
df.to_csv('flows.0.csv', index=False)
```

```python
from pathlib import Path

csv_files = sorted(Path(".").glob("converted*.csv"))

if not csv_files:
    raise FileNotFoundError("当前目录没有找到以 converted 开头的 CSV 文件。")

merged_df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index=True)
merged_df.to_csv("stand.csv", index=False)

print(f"已合并 {len(csv_files)} 个文件，输出为 stand.csv，共 {len(merged_df)} 行。")
```
